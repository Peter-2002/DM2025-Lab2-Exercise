# **Competition Code**

## 1. Preprocessing Steps

```python
import pandas as pd
import json

# Load file
with open("final_posts.json", "r",encoding="utf-8") as f:
    data = json.load(f)

# Flatten the nested JSON structure
df = pd.json_normalize(data, record_path=None)

df
```

```python
df_id = pd.read_csv("data_identification.csv")
df_id
```

```python
df_em = pd.read_csv("emotion.csv")
df_em
```

```python
#Merged DataFrame
df_posts = df[[
    'root._source.post.post_id',
    'root._source.post.text',
    'root._source.post.hashtags'
]].rename(columns={
    'root._source.post.post_id': 'id',
    'root._source.post.text': 'text',
    'root._source.post.hashtags': 'hashtags'
})

# --- 2. Merge with ID and emotion data ---
df_merged = df_posts.merge(df_id, on='id', how='left') \
                    .merge(df_em, on='id', how='left')
df_merged
```

```python
import re
def preprocess_text(text, return_tokens=False):
    if pd.isna(text) or text is None:
        return [] if return_tokens else ""
    text = text.lower()
    text = emoji.demojize(text, delimiters=(" ", " ")) 
    text = re.sub(r'http\S+|www\S+', '', text) 
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    lemmas = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]
    return lemmas if return_tokens else " ".join(lemmas)
df_merged["Processed_Text"] = df_merged["text"].apply(preprocess_text)
df_merged
```

## 2. Feature Engineering Steps

```python
#ensure hashtags is always a list
df_merged["hashtags_processed"] = df_merged["hashtags"].apply(
    lambda x: x if isinstance(x, list) else []
)
#convert hashtags list to string
df_merged["hashtags_text"] = df_merged["hashtags_processed"].apply(
    lambda x: " ".join(x) if x else ""
)
#create Processed_Text directly
df_merged["Processed_Text"] = df_merged.apply(
    lambda row: preprocess_text(row["text"]) + 
                (" " + " ".join(row["hashtags_processed"]) if row["hashtags_processed"] else ""),
    axis=1
)
df_merged = df_merged.drop(columns=["hashtags_text", "hashtags_processed", "Processed_text"], errors="ignore")
df_merged
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_merged["Emotion_ID"] = le.fit_transform(df_merged["emotion"])
label_mapping = {label: int(code) for label, code in zip(le.classes_, le.transform(le.classes_))}

print("Mapping:", label_mapping)
```

```python
#rename the column first
# TRAIN
train_df = df_merged[df_merged["split"] == "train"]
train_df = train_df.drop(
    columns=[
        "text","split","hashtags","hashtags_processed",
        "Processed_text","hashtags_text","emotion",
        "emotion_id"          # <-- remove this duplicate column
    ],
    errors="ignore"
)
# TEST
test_df = df_merged[df_merged["split"] == "test"]
test_df = test_df.drop(
    columns=[
        "text","split","emotion","hashtags","hashtags_processed",
        "Processed_text","hashtags_text","emotion_id"
    ],
    errors="ignore"
)
print(train_df.head())
print(test_df.head())
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    train_df, 
    test_size=0.1, 
    stratify=train_df["Emotion_ID"], 
    random_state=42
)

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

tfidf = TfidfVectorizer(max_features=100000, tokenizer=nltk.word_tokenize)
# Fit TF-IDF on training text
X_train = tfidf.fit_transform(train_df["Processed_Text"])
X_val   = tfidf.transform(val_df["Processed_Text"])
```

## 3. Model Implementation Steps

```python
#Add the code related to the model implementation steps in cells inside this section
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

#prepare target labels
y_train = train_df["Emotion_ID"]
y_val   = val_df["Emotion_ID"]
# max_iteration, it allows more optimized steps so model can convert
# class_weight='balanced' by giving rare classes higher weight to handle class imbalance
# solver='lbfgs' by logistic regression is supported by multinomial 
clf = LogisticRegression(max_iter=10000,class_weight='balanced',solver='lbfgs')
clf.fit(X_train, y_train)

#evaluate the model
y_pred = clf.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
X_test = tfidf.transform(test_df["Processed_Text"])

# -----------------------------------------
#         Emotional Prection ID
# -----------------------------------------
test_pred_ids = clf.predict(X_test)

id_to_emotion = {
 0: 'anger',
 1: 'disgust',
 2: 'fear',
 3: 'joy',
 4: 'sadness',
 5: 'surprise'
}

# Convert ID into emotion label
test_pred_labels = [id_to_emotion[i] for i in test_pred_ids]

# Attach to dataframe
test_df["emotion"] = test_pred_labels

# -----------------------------------------
# 4. Save as CSV in required output format
# -----------------------------------------

output_df = test_df[["id", "emotion"]]
output_df.to_csv("111072281_Emotion_Predictions.csv", index=False)

print("Saved as 111072281_Emotion_Predictions.csv")
output_df.head()
```
